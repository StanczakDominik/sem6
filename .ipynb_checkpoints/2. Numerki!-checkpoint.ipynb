## Random walk

```python
from random import random
print(random())

randomki = [random()-0.5 for i in range(1000)]
for i, randomek in enumerate(randomki[:10]):
    print(i, randomek)
print(max(randomki), min(randomki))
```

```python
import math

N = 1000000
x, y = 0, 0
x_lista, y_lista, r_lista = [], [], []

for i in range(N):
    x_lista.append(x)
    y_lista.append(y)
    r_lista.append(x**2+y**2)
    
    x += random() - 0.5
    y += random() - 0.5
    
print(r_lista)
```

# MATPLOTLIB - WASZ PRZYJACIEL
## NAWET BARDZIEJ NIŻ STALIN

```python
import matplotlib.pyplot #w cholerę długie
import matplotlib.pyplot as plt  #tak lepiej

#magiczna linijka, dosłownie tzw magic
%matplotlib inline

plt.plot(x_lista,y_lista)
plt.grid()
plt.xlabel("$x$ latexowe!")
plt.ylabel("$y$")
plt.title("Random walk!")
```

```python
fig, axes = plt.subplots(figsize=(12,8))
axes.plot(range(N),r_lista)
axes.grid()
axes.set_xlabel("Liczba iteracji")
axes.set_ylabel("Odległosc $r$")
axes.set_title("Random walk!")
```

# Random walk 2.0

```python
import math

Ncząstek = 100
N = 10000

r = np.random.random((Ncząstek, 2))
r_array = np.empty((Ncząstek, 2, N))
for i in range(N):
    r_array[:,:,i] = r
    r += np.random.random(r.shape) - 0.5  
    
fig, axes = plt.subplots(figsize=(16,10))
axes.set_xlabel("$x$")
axes.set_ylabel("$y$")
axes.set_title("Random walk 2.0")
axes.grid()
for i in range(Ncząstek):
    x = r_array[i,0,:]
    y = r_array[i,1,:]
    axes.plot(x,y, alpha=0.1)
```
